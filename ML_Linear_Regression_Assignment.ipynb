```python
new=pd.DataFrame()
# This will check and create a a dataframe with data that fall in between the upper and lower fence. 
# Basically outlier treatment. ALso remember to use operator like '|' AND '&' instead of 'or' AND 'and' because
# these don't work with pandas series
for i in df: 
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    lower_fence=Q1-(1.5*IQR)
    upper_fence=Q3+(1.5*IQR)
    new= df[(df[i]>lower_fence) & (df[i]<upper_fence)]
    #new_final[i]=new
    
    
```

# Lasso or L-1 Regularisation

```python
lassocv=LassoCV(cv=5,max_iter=2000000,normalize=True,alphas=None)
lassocv.fit(X_train,y_train)
```

```python
lassocv.alpha_
```

```python
#Fitting in Lasso Regression
lasso=Lasso(alpha=lassocv.alpha_)
lasso.fit(X_train,y_train)
```

```python
print("Training set accuracy",lasso.score(X_train,y_train))
print("Testing set accuracy",lasso.score(X_test,y_test))
```

# Ridge or L-2 Regularisation

```python
ridgecv=RidgeCV(alphas=np.random.uniform(0,10,100),cv=10,normalize=True)
ridgecv.fit(X_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge=Ridge(alpha=ridgecv.alpha_)
ridge.fit(X_train,y_train)
```

```python
print("Training set accuracy",ridge.score(X_train,y_train))
print("Testing set accuracy",ridge.score(X_test,y_test))
```
